```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
import math
import PyPDF2
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfpage import PDFPage



# creating a pdf file object
pdfFileObj = open('result.pdf','rb')

# creating a pdf reader object
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)

# printing number of pages in pdf file
num_pages = pdfReader.numPages
count = 0
text = ""


while count < num_pages:
    pageObj = pdfReader.getPage(count)
    count +=1
    text += pageObj.extractText()
   

def remove_string_special_characters(s):
    
    #replace special characters with ''
    stripped = re.sub('[^\w\s]','',s)
    stripped = re.sub('_','',stripped)
    
    #replaced any whitespace to one space
    stripped = re.sub('\s+',' ',stripped)
    
    #remove start and end white spaces
    stripped = stripped.strip()
    return stripped

#creating documents
def get_doc(sent):
    doc_info = []
    i=0
    for sent in text_sents_clean:
        i+=1
        count = count_words(sent)
        temp = {'doc_id': i, 'doc_length': count}
        doc_info.append(temp)
    return doc_info

#following functions are a pre-requisite to calculating the TF and IDF scores:
def count_words(sent):
    count = 0
    words = word_tokenize(sent)
    for word in words:
        count+=1
    return count

def create_freq_dict(sents):
    i=0
    freqDict_list = []
    for sent in sents:
        i+=1
        freq_dict = {}
        words = word_tokenize(sent)
        for word in words:
            word = word.lower()
            if word in freq_dict:
                freq_dict[word]+=1
            else:
                freq_dict[word]=1
            temp = {'doc_id': i,'freq_dict': freq_dict}
        freqDict_list.append(temp)
    return freqDict_list
    
#The functions to get the TF and IDF score:
def computeTF(doc_info, freqDict_list):

    TF_scores = []
    for tempDict in freqDict_list:
        id = tempDict['doc_id']
        for k in tempDict['freq_dict']:
            temp = {'doc_id':id,
                    'TF_scores': tempDict['freq_dict'][k]/doc_info[id-1]['doc_length'], 
                    'key':k}
            TF_scores.append(temp)
    return TF_scores

def computeIDF(doc_info, freqDict_list):
    IDF_scores = []
    counter = 0
    for dict in freqDict_list:
        counter+=1
        for k in dict['freq_dict'].keys():
            count = sum([k in tempDict['freq_dict']for tempDict in freqDict_list])
            temp = {'doc_id' : counter, 'IDF_score':math.log(len(doc_info)/count), 'key': k}
            IDF_scores.append(temp)
    return IDF_scores

def computeTFIDF(TF_scores, IDF_scores):
    TFIDF_scores = []
    for j in IDF_scores:
        for i in TF_scores:
            if j['key']==i['key'] and j['doc_id']==i['doc_id']:
                temp = {'doc_id': j['doc_id'],
                        'TFIDF_scores': j['IDF_score']*i['TF_scores'], 
                        'key': i['key']}
        TFIDF_scores.append(temp)
        #print(TFIDF_scores)
    return TFIDF_scores

            
            

            
        
```
