```python
job_location = []
for u in url:
    page = requests.get(u)
    soup = bs(page.text,'html.parser')
    container = soup.find("div", id = "job-results-page-main-content")
    jobs_container = container.find("div", class_ = "jrp-table-container")
    job = jobs_container.find_all("tr", class_="job-result")
    tags = []
    for j in range(0,len(job)):
        tag = job[j].find("a")
        tags.append(tag.get('href'))
    #print(tags)
    weblinks = []    
    for t in range(0,len(tags)):
        weblinks.append('https://careers.key.com'+tags[t])
        response  = get(weblinks[t])
        html_soup = bs(response.text,'html.parser')
        check = html_soup.find('div', class_ = 'jdp-job-snapshot-card')
        text = check.find('span', class_ = 'jobLocation').text
        job_location.append(text)
```

```python
job_title = []
for u in url:
    page = requests.get(u)
    soup = bs(page.text,'html.parser')
    container = soup.find("div", id = "job-results-page-main-content")
    jobs_container = container.find("div", class_ = "jrp-table-container")
    job = jobs_container.find_all("tr", class_="job-result")
    tags = []
    for j in range(0,len(job)):
        tag = job[j].find("a")
        tags.append(tag.get('href'))
    #print(tags)
    weblinks = []    
    for t in range(0,len(tags)):
        weblinks.append('https://careers.key.com'+tags[t])
        response  = get(weblinks[t])
        html_soup1 = bs(response.text,'html.parser')
        check1 = html_soup1.find('div', class_ = 'jdp-title-block-position')
        text1 = check1.find('span', class_ = 'jdp-title-job-title').text
        job_title.append(text1)
    print(len(weblinks))
    print(len(job_title))
    print(job_title[:10])
  
```
